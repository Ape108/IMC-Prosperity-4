# R3 Manual: Celestial Gardeners Guild

**Setup:** 51 counterparties with reserve prices uniform on {670, 675, ..., 920}. Resell value = 920.

- **Bid 1:** Trade if bid1 > reserve. PnL = (920 - bid1) per trade.
- **Bid 2:** Trade if bid2 > reserve AND bid2 > global_avg → full PnL.  
  If bid2 > reserve BUT bid2 ≤ global_avg → PnL × ((920 - avg) / (920 - bid2))³

**Historical precedent (P2R4, P3R3):** Actual global average lands only 2–3 above the Level 0 optimal.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

SELL = 920
RESERVES = np.arange(670, 921, 5)  # 51 values: 670, 675, ..., 920

def n_trades(bid: int) -> int:
    """Counterparties with reserve strictly below bid."""
    return int(np.sum(RESERVES < bid))

def penalty(avg: float, bid: int) -> float:
    """Cubic penalty multiplier when bid <= avg."""
    if bid >= SELL:
        return 0.0
    return ((SELL - avg) / (SELL - bid)) ** 3

def pnl(bid: int, avg: float) -> float:
    """Expected PnL for a single bid given the true global average."""
    trades = n_trades(bid)
    base = trades * (SELL - bid)
    if bid > avg:
        return base
    return base * penalty(avg, bid)

# --- Bid 1 sweep: pure math, no game theory ---
bids = np.arange(670, 921)
bid1_pnl = [n_trades(b) * (SELL - b) for b in bids]

best_b1 = bids[np.argmax(bid1_pnl)]
print(f"Optimal Bid 1: {best_b1}  (PnL = {max(bid1_pnl):,})")

# Show the top few to confirm the plateau
top_idx = np.argsort(bid1_pnl)[-5:][::-1]
print("\nTop Bid 1 candidates:")
for i in top_idx:
    print(f"  bid={bids[i]}  n={n_trades(bids[i])}  pnl={bid1_pnl[i]:,}")
```

## Sensitivity Analysis: Bid 2 EV vs. Assumed Global Average

For a grid of assumed true averages (790–820), plot the EV curve for Bid 2.
The vertical dashed line marks our current submission (801) and the historical
"safe just-above" zone (793–797).


```python
bid2_range = np.arange(770, 851)

# Scenarios: what might the true global average actually be?
# Historical: P2R4 avg=980 (L0=978, +2), P3R3 avg=287 (L0=284, +3)
# => P4R3 L0=791, expect avg ~ 793-795. Test a wider range for sensitivity.
avg_scenarios = [791, 793, 795, 797, 800, 805, 810, 820]

colors = cm.plasma(np.linspace(0.1, 0.85, len(avg_scenarios)))

fig, ax = plt.subplots(figsize=(14, 7))

optimal_bids = {}
for avg, color in zip(avg_scenarios, colors):
    evs = [pnl(b, avg) for b in bid2_range]
    opt_idx = np.argmax(evs)
    opt_bid = bid2_range[opt_idx]
    optimal_bids[avg] = (opt_bid, evs[opt_idx])

    label = f"avg={avg}  →  opt bid={opt_bid}"
    ax.plot(bid2_range, evs, color=color, linewidth=2, label=label)
    ax.plot(opt_bid, evs[opt_idx], 'o', color=color, markersize=7)

# Mark current submission
ax.axvline(801, color='red', linestyle='--', linewidth=1.5, label='Current Bid 2 = 801')

# Mark historical "safe zone" based on P2R4/P3R3 pattern (+2/+3 above L0)
ax.axvspan(793, 797, alpha=0.12, color='green', label='Historical safe zone (793–797)')

ax.set_title("Bid 2 Expected PnL vs. True Global Average (sensitivity)", fontsize=14)
ax.set_xlabel("Your Bid 2", fontsize=12)
ax.set_ylabel("Expected PnL", fontsize=12)
ax.legend(fontsize=10, loc='upper right')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("\nOptimal Bid 2 per assumed global average:")
print(f"{'Assumed Avg':>12} | {'Opt Bid 2':>10} | {'EV':>10}")
print("-" * 38)
for avg, (opt, ev) in optimal_bids.items():
    marker = " ◄ historical range" if avg in (793, 795) else ""
    print(f"{avg:>12} | {opt:>10} | {ev:>10,.1f}{marker}")
```

## Heatmap: EV of Bid 2 across (bid, assumed_avg) space

Shows exactly where EV is maximized and how quickly it degrades.
Red = high EV, blue = low EV. White star = current submission (801).


```python
bid2_grid = np.arange(770, 851)
avg_grid  = np.arange(785, 821)

Z = np.array([[pnl(b, a) for b in bid2_grid] for a in avg_grid])

fig, ax = plt.subplots(figsize=(14, 7))
im = ax.imshow(
    Z, aspect='auto', origin='lower',
    extent=[bid2_grid[0], bid2_grid[-1], avg_grid[0], avg_grid[-1]],
    cmap='RdYlBu_r'
)
plt.colorbar(im, ax=ax, label='Expected PnL')

# Overlay: optimal bid for each avg row
opt_bids_per_avg = [bid2_grid[np.argmax([pnl(b, a) for b in bid2_grid])] for a in avg_grid]
ax.plot(opt_bids_per_avg, avg_grid, 'w--', linewidth=2, label='Optimal bid per avg')

# Current submission
ax.plot(801, avg_grid.mean(), 'w*', markersize=18, label='Current submission (801)', zorder=5)

# Historical expected range
ax.axhspan(793, 795, alpha=0.35, color='lime', label='Historical expected avg (793–795)')
ax.axvline(795, color='lime', linestyle=':', linewidth=1.5)
ax.axvline(797, color='lime', linestyle=':', linewidth=1.5)

ax.set_title("Bid 2 EV Heatmap: (Your Bid) × (True Global Average)", fontsize=13)
ax.set_xlabel("Your Bid 2", fontsize=12)
ax.set_ylabel("True Global Average", fontsize=12)
ax.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

## Cost of Being Wrong: Penalty Cliff vs. Margin Drag

How much do you lose if:
- You're **above** the average by X units (margin drag only)
- You're **below** the average by X units (cubic penalty)

Quantifies the asymmetry that justifies bidding above expected average.


```python
true_avg = 793  # historical best estimate

# Baseline: bid exactly at the optimal if avg=793 (which is just above: 794)
base_bid = true_avg + 1
base_ev  = pnl(base_bid, true_avg)

offsets = np.arange(-15, 16)  # -15 to +15 units from true_avg
your_bids = true_avg + offsets

pct_loss = []
for b in your_bids:
    if b < 670 or b > 920:
        pct_loss.append(np.nan)
        continue
    ev = pnl(int(b), true_avg)
    pct_loss.append(100 * (ev - base_ev) / base_ev)

fig, ax = plt.subplots(figsize=(12, 6))

above_mask = offsets > 0
below_mask = offsets < 0
at_mask    = offsets == 0

ax.bar(offsets[above_mask], [p for p, m in zip(pct_loss, above_mask) if m],
       color='steelblue', alpha=0.8, label='Above avg (margin drag only)')
ax.bar(offsets[below_mask], [p for p, m in zip(pct_loss, below_mask) if m],
       color='crimson', alpha=0.8, label='Below avg (cubic penalty)')
ax.bar(offsets[at_mask], [p for p, m in zip(pct_loss, at_mask) if m],
       color='orange', alpha=0.9, label='At avg (penalty = 1.0)')

# Mark current submission's offset from assumed avg
current_offset = 801 - true_avg
if -15 <= current_offset <= 15:
    ax.axvline(current_offset, color='black', linestyle='--', linewidth=1.5,
               label=f'Current bid 801 (offset={current_offset:+d})')

ax.axhline(0, color='grey', linewidth=0.8)
ax.set_xlabel(f"Your bid offset from true avg ({true_avg})", fontsize=12)
ax.set_ylabel("% change in EV vs. bidding 1 above avg", fontsize=12)
ax.set_title(f"Asymmetry: penalty cliff (left) vs. margin drag (right)  [true avg = {true_avg}]", fontsize=13)
ax.legend(fontsize=10)
ax.grid(True, axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print(f"\nAssumed true avg: {true_avg}")
print(f"Baseline bid: {base_bid}  (EV = {base_ev:,.1f})")
print(f"\n{'Offset':>8} | {'Your Bid':>9} | {'EV':>10} | {'% vs base':>10}")
print("-" * 46)
for offset, b, loss in zip(offsets, your_bids, pct_loss):
    if np.isnan(loss):
        continue
    flag = " ◄ current" if int(b) == 801 else ""
    print(f"{offset:>+8} | {int(b):>9} | {pnl(int(b), true_avg):>10,.1f} | {loss:>+9.2f}%{flag}")
```

## Decision Summary

Change `true_avg` above to your best estimate of the crowd average and read off the table.

**Key takeaway from historical data (P2R4, P3R3):**
- True avg lands **+2 to +3** above the Level 0 optimal (791) → expect **~793–795**
- Optimal Bid 2 = avg + 1 or avg + 2 → **794–797**
- Current submission of **801** is +8 above expected avg: costs ~2–4% EV vs. a bid of 795–797
- But 801 is the right call if you believe the meta has shifted and many teams now think in levels (your partner's demographics model)

**The asymmetry rule always holds:** being 1 below the avg costs more (cubic cliff) than being 8 above (linear margin drag). So always err above, but minimally.
