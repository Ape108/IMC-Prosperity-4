```python
import numpy as np
from scipy.stats import beta, norm
```

```python
def calc_pnl(research, scale, speed, budget):
    return (research * scale * speed) - budget
```

```python
def calc_research(research_per):
    return 200_000 * np.log(1 + research_per) / np.log(1 + 100)
```

```python
def calc_scale(scale_per):
    return 7 * (scale_per / 100)
```

```python
def calc_speed(speed_per):
    def uniform_distrubution(per):
        return per / 100

    def right_skew_distrubution(per):
        return beta.cdf(per / 100, 5, 2)

    def normal_distrubution(per):
        return norm.cdf(per / 100, loc=.33, scale=.1)

    # Choose CDF based on what we think it will end up looking like 
    # Pretty sure teams with no answer (0, 0, 0) does not effect player distrubution
    # So my intuition is normal distrubution on a mean of 33 with std 10
    cdf = normal_distrubution(speed_per)
    return 0.1 + 0.8 * cdf
```

```python
def optimize():
    pillars = [0, 0, 0]
    max_pnl = 0
    for r in range(0, 100):
        for sc in range(0, 100 - r):
            for s in range(0, 100 - r - sc):
                budget = (r + sc + s) / 100 * 50_000
                pnl = calc_pnl(calc_research(r), calc_scale(sc), calc_speed(s), budget)
                if pnl >= max_pnl:
                    pillars = [r, sc, s]
                    max_pnl = pnl

    print(pillars)
    print(max_pnl)
```

```python
optimize()
```
